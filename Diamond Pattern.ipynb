# Diamond Pattern


````python
n = int(input("Please input the number of asterisks: "))

#CERMIN SEGITIGA BAGIAN ATAS

#OUTER LOOP
for i in range (n,0,-1):  
    
    #INNER LOOP
    
    #Memubat pola segitiga 1 (pojok kiri atas)
    for j in range(i,0,-1):   
        print("*", end="  ")
        
    #membuat ruang (spasi) pada segitiga 1 dan 2    
    for j in range(2*(n-i)):
        print("  ", end=" ")
    
    #membuat pola segitiga 2 (pojok kanan atas)
    for j in range (i,0,-1):
        print("*",end="  ")
        
    print()

    
#CERMIN SEGITIGA BAGIAN BAWAH
    
#OUTER LOOP
for i in range (n):
    
    #INNER LOOP
    
    #membuat segitiga 3 (pojok kiri bawah)
    for j in range(i+1):
        print("*", end="  ")
    
    #membuat ruang (spasi) pada segitiga 1 dan 2
    for j in range(2*(n-i-1)):
        print("  ", end=" ")
    
    #membuat segitiga 4 (pojok kanan bawah)
    for j in range (i+1):
        print("*",end="  ")
    
    print()
````

````python
#PEMBUKTIAN SEGITIGA 1
n = int(input("Please input the number of asterisks (Suggested to 10): "))

#CERMIN SEGITIGA BAGIAN ATAS

#OUTER LOOP
for i in range (n,0,-1):  
    
    #INNER LOOP
    
    #Memubat pola segitiga 1 (pojok kiri atas)
    for j in range(i,0,-1):   
        print(j, end="  ")
        
    #membuat ruang (spasi) pada segitiga 1 dan 22    
    for j in range(2*(n-i)):
        print("  ", end=" ")
    
    #membuat pola segitiga 2 (pojok kanan atas)
    for j in range (i,0,-1):
        print("*",end="  ")
        
    print()

#CERMIN SEGITIGA BAGIAN BAWAH
    
#OUTER LOOP
for i in range (n):
    
    #INNER LOOP
    
    #membuat segitiga 3 (pojok kiri bawah)
    for j in range(i+1):
        print("*", end="  ")
    
    #membuat ruang (spasi) pada segitiga 1 dan 2
    for j in range(2*(n-i-1)):
        print("  ", end=" ")
    
    #membuat segitiga 4 (pojok kanan bawah)
    for j in range (i+1):
        print("*",end="  ")
    
    print()

#PEMBUKTIAN SEGITIGA 2
n = int(input("Please input the number of asterisks (Suggested to 10): "))

#CERMIN SEGITIGA BAGIAN ATAS

#OUTER LOOP
for i in range (n,0,-1):  
    
    #INNER LOOP
    
    #Memubat pola segitiga 1 (pojok kiri atas)
    for j in range(i,0,-1):   
        print("*", end="  ")
        
    ##membuat ruang (spasi) pada segitiga 1 dan 22    
    for j in range(2*(n-i)):
        print("  ", end=" ")
    
    #membuat pola segitiga 2 (pojok kanan atas)
    for j in range (i,0,-1):
        print(j,end="  ")
        
    print()
    
#CERMIN SEGITIGA BAGIAN BAWAH
    
#OUTER LOOP
for i in range (n):
    
    #INNER LOOP
    
    #membuat segitiga 3 (pojok kiri bawah)
    for j in range(i+1):
        print("*", end="  ")
    
    #membuat ruang (spasi) pada segitiga 1 dan 2
    for j in range(2*(n-i-1)):
        print("  ", end=" ")
    
    #membuat segitiga 4 (pojok kanan bawah)
    for j in range (i+1):
        print("*",end="  ")
    
    print()

#PEMBUKTIAN SEGITIGA 3
n = int(input("Please input the number of asterisks (Suggested to 10): "))

#CERMIN SEGITIGA BAGIAN ATAS

#OUTER LOOP
for i in range (n,0,-1):  
    
    #INNER LOOP
    
    #Memubat pola segitiga 1 (pojok kiri atas)
    for j in range(i,0,-1):   
        print("*", end="  ")
        
    ##membuat ruang (spasi) pada segitiga 1 dan 22    
    for j in range(2*(n-i)):
        print("  ", end=" ")
    
    #membuat pola segitiga 2 (pojok kanan atas)
    for j in range (i,0,-1):
        print("*",end="  ")
        
    print()
  
#CERMIN SEGITIGA BAGIAN BAWAH
    
#OUTER LOOP
for i in range (n):
    
    #INNER LOOP
    
    #membuat segitiga 3 (pojok kiri bawah)
    for j in range(i+1):
        print(j, end="  ")
    
    #membuat ruang (spasi) pada segitiga 1 dan 2
    for j in range(2*(n-i-1)):
        print("  ", end=" ")
    
    #membuat segitiga 4 (pojok kanan bawah)
    for j in range (i+1):
        print("*",end="  ")
    
    print()
    
    
#PEMBUKTIAN SEGITIGA 4
n = int(input("Please input the number of asterisks (Suggested to 10): "))

#CERMIN SEGITIGA BAGIAN ATAS

#OUTER LOOP
for i in range (n,0,-1):  
    
    #INNER LOOP
    
    #Memubat pola segitiga 1 (pojok kiri atas)
    for j in range(i,0,-1):   
        print("*", end="  ")
        
    ##membuat ruang (spasi) pada segitiga 1 dan 22    
    for j in range(2*(n-i)):
        print("  ", end=" ")
    
    #membuat pola segitiga 2 (pojok kanan atas)
    for j in range (i,0,-1):
        print("*",end="  ")
        
    print()

#CERMIN SEGITIGA BAGIAN BAWAH
    
#OUTER LOOP
for i in range (n):
    
    #INNER LOOP
    
    #membuat segitiga 3 (pojok kiri bawah)
    for j in range(i+1):
        print("*", end="  ")
    
    #membuat ruang (spasi) pada segitiga 1 dan 2
    for j in range(2*(n-i-1)):
        print("  ", end=" ")
    
    #membuat segitiga 4 (pojok kanan bawah)
    for j in range (i+1):
        print(j,end="  ")
    
    print()
    
#TERBUKTI BOOOOOOOOOOOOSSSSSSSSSSSSSSSSSSSSSSSS
````
